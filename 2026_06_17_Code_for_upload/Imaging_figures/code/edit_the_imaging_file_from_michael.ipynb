```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import defaultdict
import numpy as np
from scipy.stats import mannwhitneyu
```

```python
# Read in the clustering data which Michael sent me
files = pd.read_csv('../data/Radius 100 - Exponential Dropoff - Self Excluded - Negative Filtered Out - with UMAP_Leiden.csv')

# Add columns called 'PID' and tp to the dataframe
files['PID'] = files['Image'].str.split('_').str[0]

# Or simpler approach
files['tp'] = files['Image'].str.extract(r'_([^_]+(?:_Progression)?)(?:\.csv|_Image)', expand=False)
files['tp'] = files['tp'].str.replace('Day_', '')

# # Delete a C6 timepoint
files = files.loc[files['tp'] != '140', :]

tp_dict = {'0': 'C1', '168': 'C7', '336': 'C12', 'Progression': 'Prog'}
files['tp'] = files['tp'].map(tp_dict)
```

```python
# Read in patient info
patient_alias = pd.read_excel('../data/patient_alias_2025.xlsx')
patient_alias_dict = dict(zip(patient_alias['PID'].astype(str), patient_alias['alias']))
```

```python
# Add additional information to the dataframe
files['PID'] = files['PID'].map(patient_alias_dict)

files['pid_tp'] = files['PID'] + '_' + files['tp']

files = files.drop(['Image'], axis = 1)
files.head()
```

```python
files.to_csv('../data/Radius_100_patient_id_edited.csv')
```
