```python
import pandas as pd
import numpy as np
import glob
import re
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
import statsmodels.api as sm
import statsmodels.formula.api as smf
from statsmodels.regression.mixed_linear_model import MixedLM
import statsmodels.stats.multitest as smm
```

```python
# Read in dataframe
cells = pd.read_csv('../data/cd34_clustering_results.csv')
cells.columns = cells.columns.str.replace('Unnamed: 0', 'FileName')
```

```python
# read in patient alias
patient_alias = pd.read_excel('../data/patient_alias_from_2025.xlsx', nrows = 40)
patient_alias['PID'].unique()
patient_alias_dict = dict(zip(patient_alias['PID'], patient_alias['alias']))
```

```python
# read in control alias
control_alias = pd.read_csv('../data/healthy_control.csv')
control_alias['value'] = control_alias['value'].str.replace('.fcs','')
control_alias['key'] = control_alias['key'].str.replace('Healthy_BMA_67y_Male','Healthy_BMA_67Y_Male')
control_alias['key'] = control_alias['key'].str.replace('_T_Cell_Panel.fcs', '')
control_alias_dict = dict(zip(control_alias['key'], control_alias['value']))
```

```python
# Create a dictionary which maps each filename to an updated filename
files_dict = {}
for ie in cells['FileName']:
    i = ie.replace('_Myeloid_Panel', '')
    if i.startswith('6'):
        x = i.split('_')[0] + i.split('_')[1]
        x = patient_alias_dict[int(x)]
        y = '_'.join(i.split('_')[2:])
        z = x + '_' + y
        files_dict[ie] = z
    elif 'Healthy' in ie:
        i = ie.replace('_Myeloid_Panel', '')
        files_dict[ie] = control_alias_dict[i]
```

```python
cells['FileName'] = cells['FileName'].map(files_dict)
```

```python
cells.to_csv('../data/cd34_clustering_results_patient_updated.csv')
```
